# Q1 (25%): This represents the first quartile, which is the value below which 25% of the data falls. For example, in the "Index" variable, 25% of the data falls below 42.75.

# Q2 (50%): This represents the second quartile, which is equivalent to the median. It divides the data into two equal halves. For example, in the "Overall Rating" variable, the median value is 4.0.

# Q3 (75%): This represents the third quartile, which is the value below which 75% of the data falls. For example, in the "Exterior" variable, 75% of the data falls below 4.9.

# 99%: This represents the 99th percentile, which indicates the value below which 99% of the data falls. For example, in the "Index" variable, 99% of the data falls below 169.29.

# Q4 (100%): This represents the maximum value in the dataset for each variable. For example, in the "Index" variable, the maximum value is 171.0.

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["1.5rule"]=1.5*descriptive[columnName]["IQR"]
    descriptive[columnName]["Lesser"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
    descriptive[columnName]["Greater"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
    descriptive[columnName]["Min"]=dataset[columnName].min()
    descriptive[columnName]["Max"]=dataset[columnName].max()
    
```

# IQR (Interquartile Range): This represents the range between the first quartile (Q1) and the third quartile (Q3). It provides a measure of the spread of the middle 50% of the data. For example:
# Index: IQR = 85.5
# Overall Rating: IQR = 4.5
# Exterior: IQR = 0.7
# Interior: IQR = 0.7
# Ride Quality: IQR = 0.6

# 1.5 * IQR rule: This rule is used to identify potential outliers in a dataset. It states that any data point that is:
# Lesser than Q1 - 1.5 * IQR or
# Greater than Q3 + 1.5 * IQRIQR

# Lesser: This represents the lower bound for identifying potential outliers using the 1.5 * IQR rule. For example:

# Index: Lesser = Q1 - 1.5 * IQR = -85.5
# Overall Rating: Lesser = Q1 - 1.5 * IQR = -6.75
# Exterior: Lesser = Q1 - 1.5 * IQR = 3.15
# Interior: Lesser = Q1 - 1.5 * IQR = 3.05
# Ride Quality: Lesser = Q1 - 1.5 * IQR = 3.3

# Greater: This represents the upper bound for identifying potential outliers using the 1.5 * IQR rule. For example:

# Index: Greater = Q3 + 1.5 * IQR = 256.5
# Overall Rating: Greater = Q3 + 1.5 * IQR = 11.25
# Exterior: Greater = Q3 + 1.5 * IQR = 5.95
# Interior: Greater = Q3 + 1.5 * IQR = 5.85
# Ride Quality: Greater = Q3 + 1.5 * IQR = 5.7

```python
def frequencytables(columnName,dataset):
    frequencytables = pd.DataFrame(columns=["unique_values", "frequency", "Relative Frequency", "cumsum"])
    frequencytables["unique_values"] =dataset[columnName].value_counts().index
    frequencytables["frequency"] = dataset[columnName].value_counts().values
    frequencytables["Relative Frequency"] = (frequencytables["frequency"] / 103)
    frequencytables["cumsum"] = frequencytables["Relative Frequency"].cumsum()
    return frequencytables
```

```python
def calculate_descriptive_statistics(dataset, quan):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                                   "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max","skew","kurtosis","var","std"],columns=quan)
    for columnName in quan:
        descriptive[columnName]["Mean"]=dataset[columnName].mean()
        descriptive[columnName]["Median"]=dataset[columnName].median()
        descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
        descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
        descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
        descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
        descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
        descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
        descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
        descriptive[columnName]["1.5rule"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["Lesser"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
        descriptive[columnName]["Greater"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
        descriptive[columnName]["Min"]=dataset[columnName].min()
        descriptive[columnName]["Max"]=dataset[columnName].max()
        descriptive[columnName]["skew"]=dataset[columnName].skew()
        descriptive[columnName]["kurtosis"]=dataset[columnName].kurtosis()
        descriptive[columnName]["var"]=dataset[columnName].var()
        descriptive[columnName]["std"]=dataset[columnName].std()
    return descriptive
```

# Since the p-value (0.016) is less than the commonly used significance level of 0.05, we reject the null hypothesis.

# The p-value (in this case, approximately 0.016) is the probability of observing the given result (or a more extreme result) if the null hypothesis is true. In hypothesis testing, the null hypothesis typically assumes that there is no difference between the groups being compared. A low p-value indicates that there is strong evidence against the null hypothesis, suggesting that the observed difference between the groups is statistically significant.

```python
from scipy.stats import ttest_ind

# Filter the dataset for rows where "Overall Rating" equals 4.4 (just as an example)
car_rating = dataset[dataset["Overall Rating"] == 4.4]["Exterior"]
car_rating1 = dataset[dataset["Overall Rating"] == 4.4]["Interior"]
# Perform the t-test
ttest_ind(car_rating, car_rating1)
```
